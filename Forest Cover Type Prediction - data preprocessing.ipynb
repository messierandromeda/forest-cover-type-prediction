**Forest Cover Type Prediction - preprocessing the data**

```python
# importing the required packages

import numpy as np
import pandas as pd
import os

path = os.getcwd()   # current working directory
current_folder = "data"
print(os.getcwd())
for dirname, _, filenames in os.walk(os.path.join(path,current_folder)):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
# Loading training data
train_data = pd.read_csv(os.path.join(path,"data/train.csv"))
train_data.head()
```

```python
train_data.info()     # there are no missing data
```

```python
train_data.describe()
```

```python
# Combine the columns Wilderness_Area1, ..., Wilderness_Area4 to a single column, because the values are mutually exclusive
# Do the same for the columns Soil_Type1, ..., Soil_Type 40
N = len(train_data)

arr = ["Soil_Type"+str(i) for i in range(1,41)]  # all the columns in Soil_Type
#print(arr)

arr1 = ["Wilderness_Area"+str(j) for j in range(1,5)]  # all the columns in Wilderness_Area
#print(arr1)

soil_type = [0]*N   # create a new array with Soil_Type as a number 

# Soil_Type1 to 40
for j in range(N):  # total amount of training data
    for i in range(40):  # number of soil_type classes
        if train_data[arr[i]][j] == 1:   # if there is a 1 in a column between Soil_Type1 and Soil_Type40
            soil_type[j] = i+1             # add the corresponding number to the new array
#print("New column for Soil_Type1 to 40:", soil_type)

# do the same for Wilderness_Area1 to 4
wilderness_area = [0]*N

for j in range(N):
    for i in range(4):
        if train_data[arr1[i]][j] == 1:
            wilderness_area[j] = i+1
#print("New column for Wilderness_Area1 to 4:", wilderness_area)
```

```python
# Loading testing data
test_data = pd.read_csv(os.path.join(path,"data/test.csv"))
test_data.head()
```

```python
test_data.describe()
```

```python
# do the same for the test data

n = len(test_data)
print(n)
arr_ = ["Soil_Type"+str(i) for i in range(1,41)]  # all the columns in Soil_Type
arr1_ = ["Wilderness_Area"+str(j) for j in range(1,5)]  # all the columns in Wilderness_Area

soiltype = [0]*n   # create a new array with Soil_Type as a number 

# Soil_Type1 to 40
for j in range(n):  # total amount of testing data
    for i in range(40):  # number of soil_type classes
        if test_data[arr_[i]][j] == 1:   # if there is a 1 in a column between Soil_Type1 and Soil_Type40
            soiltype[j] = i+1             # add the corresponding number to the new array
#print("New column for Soil_Type1 to 40:", soiltype)

# do the same for Wilderness_Area1 to 4
wildernessarea = [0]*n

for j in range(n):
    for i in range(4):
        if test_data[arr1_[i]][j] == 1:
            wildernessarea[j] = i+1
#print("New column for Wilderness_Area1 to 4:", wildernessarea)
```

```python
# export the new columns to a new file for the training data
output = pd.DataFrame({"Id": train_data.Id, "Elevation": train_data.Elevation, 
                       "Aspect": train_data.Aspect, "Slope": train_data.Slope, 
                       "Horizontal_Distance_To_Hydrology": train_data.Horizontal_Distance_To_Hydrology, 
                       "Vertical_Distance_To_Hydrology": train_data.Vertical_Distance_To_Hydrology, 
                       "Horizontal_Distance_To_Roadways": train_data.Horizontal_Distance_To_Roadways, 
                       "Hillshade_9am": train_data.Hillshade_9am, "Hillshade_Noon": train_data.Hillshade_Noon, 
                       "Hillshade_3pm": train_data.Hillshade_3pm, 
                       "Horizontal_Distance_To_Fire_Points": train_data.Horizontal_Distance_To_Fire_Points,
                       "Wilderness_Area": wilderness_area, "Soil_Type": soil_type, "Cover_Type": train_data.Cover_Type})
output.to_csv('data/new_train_data.csv', index=False) 
```

```python
# export the new columns to a new file for the testing data
output = pd.DataFrame({"Id": test_data.Id, "Elevation": test_data.Elevation, 
                       "Aspect": test_data.Aspect, "Slope": test_data.Slope, 
                       "Horizontal_Distance_To_Hydrology": test_data.Horizontal_Distance_To_Hydrology, 
                       "Vertical_Distance_To_Hydrology": test_data.Vertical_Distance_To_Hydrology, 
                       "Horizontal_Distance_To_Roadways": test_data.Horizontal_Distance_To_Roadways, 
                       "Hillshade_9am": test_data.Hillshade_9am, "Hillshade_Noon": test_data.Hillshade_Noon, 
                       "Hillshade_3pm": test_data.Hillshade_3pm, 
                       "Horizontal_Distance_To_Fire_Points": test_data.Horizontal_Distance_To_Fire_Points,
                       "Wilderness_Area": wildernessarea, "Soil_Type": soiltype})
output.to_csv('data/new_test_data.csv', index=False) 
```
